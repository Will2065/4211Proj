```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
Data_train=pd.read_csv('Data-train.csv',index_col=0)
```

```python
feature={}
feature["attribute1"]=Data_train.attribute1
feature["attribute2"]=Data_train.attribute2
data=pd.DataFrame(feature)
```

K=3

```python
clf = KMeans(n_clusters=3)
clf.fit(data)  # 分组
centers = clf.cluster_centers_ # 两组数据点的中心点
labels = clf.labels_   # 每个数据点所属分组
print(centers)
print(labels) 
Data_train['Labels_3'] =labels
Data_train.to_csv(r"Data-train.csv",mode = 'a',index =False)
```

```python
Data_train.head()
```

```python
for i in range(len(labels)):
    
    colour=''
    if labels[i] == 0:
        colour='r'
    elif labels[i]==1:
        colour='b' 
    elif labels[i]==2:
        colour='m'
    plt.scatter(data.attribute1[i], data.attribute2[i], c=colour)    
plt.scatter(centers[:,0],centers[:,1],marker='*', s=100)
plt.show()
```
